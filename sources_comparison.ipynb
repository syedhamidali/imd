```python
import cartopy.crs as ccrs
import cartopy.feature as cfeat
import matplotlib.pyplot as plt 
```

```python
fig = plt.figure(figsize=[10,10])
ax = plt.axes(projection = ccrs.PlateCarree())
ax.set_extent([65, 100, 5, 37])
ax.add_feature(cfeat.BORDERS, alpha = 0.5, lw = 0.5)
ax.add_feature(cfeat.COASTLINE, alpha = 0.5, lw = 0.5)
ax.add_feature(cfeat.STATES.with_scale("10m"), alpha = 0.5, lw = 0.5)
ax.add_feature(cfeat.LAND, alpha = 0.5)
ax.add_feature(cfeat.OCEAN, alpha = 0.5)
df.plot(x = "lon", y = "lat", kind = "scatter", fc = "none", 
        ec = "k", s = 200, ax = ax, label = "Multi Sources")

df2.plot(x = "lon", y = "lat", kind = "scatter", fc = "none", 
         ec = "r", s = 200, ax = ax, label = "Doyle et. al., 2016")
```
